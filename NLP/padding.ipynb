## 패딩(padding)

## 1. Numpy로 패딩하기

텍스트 데이터 세팅

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
preprocessed_sentences = [['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]
```

vocab 생성 & 정수 인코딩

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(preprocessed_sentences) # vocab 생성
encoded = tokenizer.texts_to_sequences(preprocessed_sentences) # 정수 인코딩
print(encoded)
```

가장 긴 문장의 길이 계산

```python
max_len = max(len(item) for item in encoded)
print('최대 길이 :', max_len)
```

가장 긴 문장의 길이로 모든 문장을 패딩

```python
for sentence in encoded:
  while len(sentence) < max_len:
    sentence.append(0)

padded_np = np.array(encoded)
padded_np
```

## 2. 케라스 전처리 도구로 패딩하기

케라스의 전처리 도구 pad_sequences()

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

encoded를 패딩 이전의 값으로 다시 되돌려줌

```python
encoded = tokenizer.texts_to_sequences(preprocessed_sentences)
print(encoded)
```

pad_sequences 사용하여 패딩

```python
padded = pad_sequences(encoded)
padded
```

문장 뒤에 0을 채우고 싶다면 padding=’post’ 인자 추가

```python
padded = pad_sequences(encoded, padding='post')
padded
```

Numpy를 이용한 패딩 결과와 비교

```python
((padded) == (padded_np)).all()
```

패딩 길이 지정

```python
padded = pad_sequences(encoded, padding='post', maxlen=5)
padded
```

데이터가 손실될 경우 앞의 단어가 아니라 뒤의 단어라 삭제되도록 하고 싶다면 truncating=’post’ 사용

```python
padded = pad_sequences(encoded, padding='post', maxlen=5, truncating='post')
padded
```

0이 아닌 다른 숫자로 패딩할 경우

```python
last_value = len(tokenizer.word_index)+1 # 단어 집합의 크기보다 1 큰 숫자를 사용
print(last_value)
```

```python
padded = pad_sequences(encoded, padding='post', value=last_value)
padded
```

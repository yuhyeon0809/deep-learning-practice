## 데이터의 분리

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## X와 y 분리하기

---

### 1) `zip()` 함수를 이용하여 분리

`zip()` 함수는 *동일한 개수*를 가지는 시퀀스 자료형에서 각 순서에 등장하는 원소들끼리 묶어주는 역할을 한다. 

```python
X, y = zip(['a', 1], ['b', 2], ['c', 3])
print('X 데이터 :', X)
print('y 데이터 :', y)
```

```python
# 리스트의 리스트 또는 행렬 또는 뒤에서 배울 개념인 2D 텐서.
sequences = [['a', 1], ['b', 2], ['c', 3]]
X, y = zip(*sequences)
print('X 데이터 :', X)
print('y 데이터 :', y)
```

### 2) 데이터프레임을 이용하여 분리

```python
values = [['당신에게 드리는 마지막 혜택!', 1],
['내일 뵐 수 있을지 확인 부탁드...', 0],
['도연씨. 잘 지내시죠? 오랜만입...', 0],
['(광고) AI로 주가를 예측할 수 있다!', 1]]
columns = ['메일 본문', '스팸 메일 유무']

df = pd.DataFrame(values, columns=columns)
df
```

데이터프레임은 열 이름으로 각 열에 접근 가능하므로, 이를 이용해 손쉽게 X와 y를 분리할 수 있다. 

```python
X = df['메일 본문']
y = df['스팸 메일 유무']

print('X 데이터 :', X.to_list())
print('y 데이터 :', y.to_list())
```

### 3) Numpy를 이용하여 분리

`Numpy`의 `slicing`을 사용하여 분리

```python
np_array = np.arange(0,16).reshape((4,4))
print('전체 데이터 :')
print(np_array)
```

마지막 열을 제외하고 X에 저장, 마지막 열만 y에 저장

```python
X = np_array[:, :3]
y = np_array[:,3]

print('X 데이터 :')
print(X)
print('y 데이터 :', y)
```

## 테스트 데이터 분리하기

---

### 1) `sklearn` 이용하여 분리

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

- `X` : 독립 변수 데이터 (배열 또는 데이터프레임)
- `y` : 종속 변수 데이터 (레이블)
- `test_size` : 테스트용 데이터의 개수 지정. 1보다 작은 실수일 경우 비율을 나타냄
- `train_size` : 학습용 데이터의 개수 지정. 1보다 작은 실수일 경우 비율을 나타냄
- `random_state` : 난수 seed값
  - `train_test_split()`은 기본적으로 데이터의 순서를 섞고나서 분리하기 때문에 필요

```python
# 임의로 X와 y 데이터를 생성
X, y = np.arange(10).reshape((5,2)), range(5)

print('X 전체 데이터 :')
print(X)
print('y 전체 데이터 :')
print(list(y))
```

> 실행 시 `random_state` 값을 동일하게 기재해주면 항상 동일한 훈련 데이터와 테스트 데이터를 얻을 수 있다. 하지만 이 값을 변경하면 다른 순서로 섞인 채 분리되므로 이전과 다른 훈련 데이터와 테스트 데이터를 얻는다. 

```python
# 7:3의 비율로 훈련 데이터와 테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
print('X 훈련 데이터 :')
print(X_train)
print('X 테스트 데이터 :')
print(X_test)
```

```python
print('y 훈련 데이터 :')
print(y_train)
print('y 테스트 데이터 :')
print(y_test)
```

`random_state` 값을 임의의 다른 값인 `1`로 주고 다시 분리

```python
# random_state의 값을 변경
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
print('y 훈련 데이터 :')
print(y_train)
print('y 테스트 데이터 :')
print(y_test)
```

`random_state` 값이 `1234`일 때와 전혀 다른 `y` 데이터가 출력된다. 

### 2) 수동으로 분리

```python
# 실습을 위해 임의로 X와 y가 이미 분리 된 데이터를 생성
X, y = np.arange(0,24).reshape((12,2)), range(12)

print('X 전체 데이터 :')
print(X)
print('y 전체 데이터 :')
print(list(y))
```

훈련 데이터의 개수와 테스트 데이터의 개수 정하기

```python
num_of_train = int(len(X) * 0.8) # 데이터의 전체 길이의 80%에 해당하는 길이값을 구한다.
num_of_test = int(len(X) - num_of_train) # 전체 길이에서 80%에 해당하는 길이를 뺀다.

print('훈련 데이터의 크기 :', num_of_train)
print('테스트 데이터의 크기 :', num_of_test)
```

> 여기서 `num_of_test`를 `len(X) * 02` 로 계산해서는 안된다. 
**데이터 누락**이 발생할 수 있기 때문





```python
X_test = X[num_of_train:] # 전체 데이터 중에서 20%만큼 뒤의 데이터 저장
y_test = y[num_of_train:] # 전체 데이터 중에서 20%만큼 뒤의 데이터 저장
X_train = X[:num_of_train] # 전체 데이터 중에서 80%만큼 앞의 데이터 저장
y_train = y[:num_of_train] # 전체 데이터 중에서 80%만큼 앞의 데이터 저장
```

> 데이터를 나눌 때는 `num_of_train`과 같이 하나의 변수만 사용하면 데이터 누락을 방지할 수 있다. 

```python
print('X 테스트 데이터 :')
print(X_test)
print('y 테스트 데이터 :')
print(list(y_test))
```

> `train_test_split()` 과 다른 점은 데이터를 섞지 않은 채 어느 지점에서 데이터를 앞과 뒤로 분리했다는 것이다. 만약, 수동으로 데이터 분리를 진행할 경우 데이터를 분리하기 전에 수동으로 데이터를 섞는 과정이 필요할 수 있다.
